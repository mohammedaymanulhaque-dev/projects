Declare feature vector and target variable

```python
X = df.drop(['Temperature'], axis=1)

y = df['Temperature']
```

Split data into separate training and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=['Revenue'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_test.head()
```

Random Forest Classifier model with default parameters

```python
# import Random Forest regressor
from sklearn.ensemble import RandomForestRegressor

# instantiate the regressor
rfr = RandomForestRegressor(random_state=0)

# fit the model
rfr.fit(X_train, y_train)

# Predict the Test set results
y_pred = rfr.predict(X_test)

# Check evaluation metric (e.g., Mean Absolute Error)
from sklearn.metrics import mean_absolute_error

print('Model Mean Absolute Error with 10 decision-trees :', mean_absolute_error(y_test, y_pred))
```

Random Forest Classifier model with parameter n_estimators=100

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

rfr_100 = RandomForestRegressor(n_estimators=100, random_state=0)

# fit the model to the training set
rfr_100.fit(X_train, y_train)

# Predict on the test set results
y_pred_100 = rfr_100.predict(X_test)

# Check evaluation metric (e.g., Mean Absolute Error)
print('Model Mean Absolute Error with 100 decision-trees :', mean_absolute_error(y_test, y_pred_100))
```

```python
# create the regressor with n_estimators = 100
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=100, random_state=0)

# fit the model to the training set
rfr.fit(X_train, y_train)
```

```python
# view the feature scores

feature_scores = pd.Series(rfr.feature_importances_, index=X_train.columns).sort_values(ascending=False)

feature_scores
```
