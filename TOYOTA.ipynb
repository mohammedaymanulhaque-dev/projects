**Correlation matrix**

**Data Preparation and Pre-processing**

```python
df = df.dropna()
```

Convert categorical variables to numerical (example with 'Fuel_Type

```python
df = pd.get_dummies(df, columns=['Fuel_Type'], drop_first=True)
```

 Normalize numerical features

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_features = ['Age_08_04', 'KM', 'HP', 'Quarterly_Tax', 'Weight']
df[numerical_features] = scaler.fit_transform(df[numerical_features])

print("\nFirst few rows after pre-processing:")
print(df.head())
# Save the preprocessed dataset
df.to_csv('toyota_corolla_preprocessed.csv', index=False)

print("\nPre-processing complete. Preprocessed dataset saved as 'toyota_corolla_preprocessed.csv'")
```
